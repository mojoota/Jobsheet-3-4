```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat array n dari 0 sampai 15
n = np.arange(16)

# Membuat respons impuls h dari filter dengan formula sinc
h = np.sin(0.14 * np.pi * (n - 7.5)) / (np.pi * (n - 7.5))

# Membuat jendela Kaiser hk dengan panjang 16 dan faktor beta 8
hk = np.kaiser(16, 8)

# Melakukan konvolusi antara h dan hk untuk mendapatkan respons impuls filter yang difilter
hfilt = hk * h

# Membuat gambar baru untuk plot dengan ukuran yang ditentukan
plt.figure(figsize=(10, 8))

# Memplot respons impuls filter yang difilter
plt.plot(hfilt)

# Menambahkan garis grid pada plot
plt.grid()

# Menampilkan plot
plt.show()
```

Gambar di atas memperlihatkan sebuah grafik berbentuk kurva lonceng yang umumnya dikenal sebagai distribusi normal atau Gaussian. Sumbu horizontal (sumbu x) pada grafik ini tampaknya menunjukkan suatu variabel yang berubah-ubah, sedangkan sumbu vertikal (sumbu y) mungkin merupakan kepadatan probabilitas atau frekuensi dari variabel tersebut. Puncak kurva terletak di sekitar nilai 8 pada sumbu x, yang menunjukkan bahwa nilai ini adalah nilai rata-rata atau median dari data yang diwakilkan. Kurva ini simetris dengan ekor yang mengecil ke arah nilai lebih rendah dan lebih tinggi dari mean, menunjukkan penyebaran data yang standar dalam distribusi normal, dengan standar deviasi yang mengontrol lebar puncak. Kurva ini banyak digunakan dalam statistik untuk mewakili data yang alaminya mengikuti pola sebaran yang teratur dan prediktif.


```python
import scipy.signal as signal

def freqz(x):
    # Mendapatkan bins frekuensi dan nilai magnitudo sesuai
    f, H = signal.freqz(x)

    # Frekuensi dinormalisasi
    fNorm = f / np.pi

    # Magnitudo dalam dB
    hdB = 20 * np.log10(abs(H) + 1e-5)  # "+1e-5" menghindari log10(0)!

    # Membuka gambar
    plt.figure(figsize=(10, 8))

    # Plot untuk magnitudo
    plt.subplot(211)
    plt.title('Magnitudo')
    plt.plot(f, hdB, 'b')
    plt.axis([0, np.pi, np.min(hdB) - 1, np.max(hdB) + 1])
    plt.grid(True)
    plt.xlabel('Frekuensi Ter-Normalisasi (rad/sample)')
    plt.ylabel('Magnitudo (dB)')

    # Plot untuk fase
    angles = np.unwrap(np.angle(H))
    anglesGrad = (360 * angles) / (2 * np.pi)
    plt.subplot(212)
    plt.title('Fase')
    plt.plot(f, anglesGrad, 'b')
    plt.axis([0, np.pi, np.min(anglesGrad), np.max(anglesGrad)])
    plt.grid(True)
    plt.xlabel('Frekuensi Ter-Normalisasi (rad/sample)')
    plt.ylabel('Fase (derajat)')
    plt.tight_layout()
```

```python
# Memanggil fungsi 'freqz' dengan parameter 'hfilt', yang merupakan respons impuls dari filter yang difilter
freqz(hfilt)
```
